```python
import pandas as pd

geral = pd.read_csv(r"./unprocessed_musics.csv")

geral.head()
```

```python
# Separate in two datasets: one for track and another for related tracks

tracks = geral[['artist', 'track', 'track_url', 'tags']]

related_tracks = geral[['related_artist', 'related_track', 'related_track_url', 'tags']]
related_tracks.columns = ['artist', 'track', 'track_url', 'tags']
```

```python
# Concatenate datasets

tracks = pd.concat([tracks, related_tracks], ignore_index=True)
tracks = tracks.drop_duplicates(subset=['artist', 'track', 'track_url']).dropna()
tracks.head()

tracks.to_csv('tracks_df.csv', index=False)
```
